```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def julia(z, c, maxiter):
    for iteration in range(maxiter):
        z = (z*z) + c
        if abs(z) > 4:
            break
            pass 
        pass
    return iteration
```

```python
def show_plot(atlas):
    fig = plt.figure(figsize = (18,18))
    ax = fig.add_subplot(111)
    ax.imshow(atlas.T, interpolation='nearest')
```

```python
def get_atlas(x_list,y_list,c):
    xlen = len(x_list)
    ylen = len(y_list)

    #Reset the atlas
    atlas = np.empty([len(x_list),len(y_list)])

    for ix in range(xlen):
        for iy in range(ylen):
            z = complex(x_list[ix],y_list[iy])
            atlas[ix,iy] = julia(z,c,80)
            pass
        pass
    return atlas
```

## The start

```python
# value of c (unique for each Julia set)
c = complex(-0.45, 0.50)

# set the location and size of the atlas rectangle
x_list = np.linspace(-2, 2, 1000)
y_list = np.linspace(-2, 2, 1000)

atlas = get_atlas(x_list,y_list,c)

show_plot(atlas)
```
