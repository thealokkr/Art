```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Mandelbrot function. C is the constant complex number from the equation. Maxiter is the number of iteration
#to decide on the divergence
def mandel(c, maxiter):
    z = complex(0,0)

    #For now if the magnitude is greter than 4 we will assume it to be diverging. Different iteration rates also
    #offers lot of creative oppurtunities
    for iteration in range(maxiter):
        z = (z*z) + c
        if abs(z) > 4:
            break
            pass 
        pass
    return iteration
```

```python
def show_plot(atlas):
    fig = plt.figure(figsize = (18,18))
    ax = fig.add_subplot(111)
    ax.imshow(atlas.T, interpolation='nearest')
```

```python
def get_atlas(x_list,y_list,c, maxIter):
    
    #Create an empty atlas
    xlen = len(x_list)
    ylen = len(y_list)
    
    atlas = np.empty([xlen,ylen])

    for ix in range(xlen):
        for iy in range(ylen):
            c = complex(x_list[ix],y_list[iy])
            atlas[ix,iy] = mandel(c,maxIter)
            pass
        pass
    
    return atlas
```

## The start

```python
Full view mandlebrot
# value of c
c = complex(-0.45, 0.50)

# set the location and size of the atlas rectangle
x_list = np.linspace(-2.25, 0.75, 1000)
y_list = np.linspace(-1.5, 1.5, 1000)

atlas = get_atlas(x_list,y_list,c,80)

show_plot(atlas)
```
